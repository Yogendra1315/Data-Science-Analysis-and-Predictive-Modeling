```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame({
    'country':np.random.choice(['Aus', 'India', 'Uk', 'Usa'], 20),
    'gender':np.random.choice(['male', 'female'], 20),
    'salary':np.random.randint(30, 90, 20),
    'height':np.random.randint(120, 190, 20),
    'weight':np.random.randint(40, 120, 20)
}, index=[chr(65+i) for i in range(20)])
```
