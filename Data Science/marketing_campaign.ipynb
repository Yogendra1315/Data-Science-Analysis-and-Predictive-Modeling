```python
df=sns.load_dataset('taxis')
df
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.scatter(df['distance'],df['fare'])
plt.xlabel('Distance')
plt.ylabel('Fare')
plt.show()
```

```python
df['color'].value_counts()
```

```python
color=df['color'].map({
    'yellow':'y',
    'green':'g'
})
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.scatter(df['distance'],df['fare'],c=color)
plt.xlabel('Distance')
plt.ylabel('Fare')
plt.show()
```

```python
df['payment'].value_counts()
```

```python
size=df['payment'].map({
    'credit card':20,
    'cash':40
})
```

```python
plt.figure(figsize=(16,9),dpi=200)
plt.scatter(df['distance'],df['fare'],c=color,s=size)
plt.xlabel('Distance')
plt.ylabel('Fare')
plt.show()
```
